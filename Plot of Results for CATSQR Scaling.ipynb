```python
import numpy as np
import matplotlib.pyplot as plt

# For fixed columns experiments (fixed columns = 3):
# x values represent the number of rows.
# Rows start at 48 and increase in multiples of 48.
x_fixed_columns = np.array([48, 96, 144, 192, 240])

# Weak scaling results (execution times in seconds) for fixed columns,
# with different thread counts 1, 2, 4, and 8.
w1 = [0.000585, 0.000708, 0.000578, 0.000657, 0.000646]
w2 = [0.000467, 0.000484, 0.000369, 0.000514, 0.000569]
w4 = [0.000439, 0.000531, 0.000482, 0.000547, 0.000527]
w8 = [0.000610, 0.000714, 0.000731, 0.000975, 0.000677]

# Strong scaling results for fixed columns.
s1 = [0.000439, 0.000255, 0.000902, 0.000827, 0.000915]
s2 = [0.006677, 0.000256, 0.000524, 0.000418, 0.000744]
s4 = [0.000671, 0.000267, 0.001190, 0.000565, 0.000565]
s8 = [0.000888, 0.003377, 0.000549, 0.000522, 0.000594]

# For fixed rows experiments (fixed rows = 800):
# x values represent the number of columns, starting at 10 and increasing by 10.
x_fixed_rows = np.array([10, 20, 30, 40, 50])

# Weak scaling results for fixed rows.
wc1 = [0.001170, 0.002460, 0.004146, 0.005657, 0.007256]
wc2 = [0.001086, 0.001772, 0.003286, 0.003928, 0.004310]
wc4 = [0.001150, 0.002616, 0.002911, 0.004524, 0.009707]
wc8 = [0.002342, 0.003376, 0.006029, 0.011239, 0.014080]

# Strong scaling results for fixed rows.
sc1 = [0.002395, 0.005126, 0.007896, 0.012164, 0.016197]
sc2 = [0.001272, 0.002312, 0.003272, 0.006340, 0.007474]
sc4 = [0.001704, 0.001581, 0.003198, 0.004994, 0.008245]
sc8 = [0.001137, 0.001942, 0.002783, 0.005472, 0.006499]

# Two subplots side-by-side.
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Subplot 1: Fixed columns experiments (rows vary)
ax = axs[0]
ax.plot(x_fixed_columns, w1, marker='o', linestyle='-', label='Weak scaling, 1 thread')
ax.plot(x_fixed_columns, w2, marker='s', linestyle='-', label='Weak scaling, 2 threads')
ax.plot(x_fixed_columns, w4, marker='^', linestyle='-', label='Weak scaling, 4 threads')
ax.plot(x_fixed_columns, w8, marker='d', linestyle='-', label='Weak scaling, 8 threads')

ax.plot(x_fixed_columns, s1, marker='o', linestyle='--', label='Strong scaling, 1 thread')
ax.plot(x_fixed_columns, s2, marker='s', linestyle='--', label='Strong scaling, 2 threads')
ax.plot(x_fixed_columns, s4, marker='^', linestyle='--', label='Strong scaling, 4 threads')
ax.plot(x_fixed_columns, s8, marker='d', linestyle='--', label='Strong scaling, 8 threads')

ax.set_xlabel('Number of Rows')
ax.set_ylabel('Execution Time (s)')
ax.set_title('Fixed Columns (3): Scaling vs. Row Size')
ax.legend()
ax.grid(True)

# Subplot 2: Fixed rows experiments (columns vary)
ax = axs[1]
ax.plot(x_fixed_rows, wc1, marker='o', linestyle='-', label='Weak scaling, 1 thread')
ax.plot(x_fixed_rows, wc2, marker='s', linestyle='-', label='Weak scaling, 2 threads')
ax.plot(x_fixed_rows, wc4, marker='^', linestyle='-', label='Weak scaling, 4 threads')
ax.plot(x_fixed_rows, wc8, marker='d', linestyle='-', label='Weak scaling, 8 threads')

ax.plot(x_fixed_rows, sc1, marker='o', linestyle='--', label='Strong scaling, 1 thread')
ax.plot(x_fixed_rows, sc2, marker='s', linestyle='--', label='Strong scaling, 2 threads')
ax.plot(x_fixed_rows, sc4, marker='^', linestyle='--', label='Strong scaling, 4 threads')
ax.plot(x_fixed_rows, sc8, marker='d', linestyle='--', label='Strong scaling, 8 threads')

ax.set_xlabel('Number of Columns')
ax.set_ylabel('Execution Time (s)')
ax.set_title('Fixed Rows (800): Scaling vs. Column Size')
ax.legend()
ax.grid(True)

plt.suptitle('Strong and Weak Scaling Results for CA-TSQR')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

Varying the column size appears to lead to slower execution time in both the weak and strong scaling case, but varying row size for fixed column size appears to have no scaling effect. As n (column dimension) grows, the arithmetic cost of each QR step grows quadratically.
